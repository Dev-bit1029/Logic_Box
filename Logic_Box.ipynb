```python
for i in range(1, 6):
    print("*" * i)
```

```python

i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
n = 5

for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)
```

```python
n = 5

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
n = 5

# Upper half
for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)

# Lower half
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "* " * i)
```

```python
i = -5

while i <= 5:
    if i > 0:
        print(i, "Positive")
    elif i < 0:
        print(i, "Negative")
    else:
        print(i, "Zero")
    
    i += 1
```

```python
start = 10
end = 15

for i in range(start, end + 1):
    if i % 2 == 0:
        print(i, "is Even")
    else:
        print(i, "is Odd")
```

```python
i = 1

while i <= 30:
    if i % 2 == 0 and i % 3 == 0 and i % 5 == 0:
        print(i, "Divisible by 2, 3 and 5")
    elif i % 2 == 0 and i % 3 == 0:
        print(i, "Divisible by 2 and 3")
    elif i % 2 == 0:
        print(i, "Divisible by 2")
    elif i % 3 == 0:
        print(i, "Divisible by 3")
    elif i % 5 == 0:
        print(i, "Divisible by 5")
    else:
        print(i, "None")
    
    i += 1
```

```python
i = 1

while i <= 50:
    s = sum(int(d) for d in str(i))
    
    if s % 9 == 0:
        print(i, "Divisible by 9")
    else:
        print(i)
    
    i += 1
```

```python
i = 1
while i <= 20:
    if i % 2 != 0:
        print(i)
    i += 1
```

```python
for i in range(3):        # outer loop
    for j in range(3):    # inner loop
        print(i, j)
```

```python
n = 5
num = 1

for i in range(1, n+1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
n = 5
for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end="")
    for k in range(2*i-1):
        print("1", end="")
    print()
```
